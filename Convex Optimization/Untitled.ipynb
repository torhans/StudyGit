```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(-10,10,1000)
x100=np.linspace(-100,0,1000)
```

```python
def h(x):
    return (x)**2

def g(x):
    return -x**2
```

```python
plt.plot(x100,h(x100))
```

```python
plt.plot(x,g(x))
```

```python
plt.plot(x,h(g(x)))
```
