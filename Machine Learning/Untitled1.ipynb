```python
import scipy as sc
```

```python
def fun(w,x,y,lm):
    err=(np.sum(np.square(np.matmul(x,w)-y)))
    pen=lm*np.sum(np.abs(w))
    return(err+pen)
```

```python
res=sc.optimize.minimize(fun,res.x,(x,y,.5))
```
