```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
c=[1,2]
f=(c[0]+c[1])/2
err=f**2-1-f
n=0
while np.abs(err)>1e-15:
    n+=1
    if err>0:
        c[1]=f
    else:
        c[0]=f
    f=(c[0]+c[1])/2
    err=f**2-1-f
    if n>1000:
        break
n,err,f
```

```python
%matplotlib inline
plt.figure(figsize=(20,20))
n=50
ang=np.pi/2/n
k=1/f**(1/n)
R=np.array([[k*np.cos(ang),-k*np.sin(ang)],[k*np.sin(ang),k*np.cos(ang)]])
v=[1,0]
vx=[v[0]]
vy=[v[1]]
for i in range(20*n):
    v=np.matmul(R,v)
    vx.append(v[0])
    vy.append(v[1])
plt.plot(vx,vy)

v=[np.sqrt(1/2),np.sqrt(1/2)]
vx=[v[0]]
vy=[v[1]]
for i in range(20*n):
    v=np.matmul(R,v)
    vx.append(v[0])
    vy.append(v[1])
plt.plot(vx,vy)

v=[0,1]
vx=[v[0]]
vy=[v[1]]
for i in range(20*n):
    v=np.matmul(R,v)
    vx.append(v[0])
    vy.append(v[1])
plt.plot(vx,vy)

v=[-np.sqrt(1/2),np.sqrt(1/2)]
vx=[v[0]]
vy=[v[1]]
for i in range(20*n):
    v=np.matmul(R,v)
    vx.append(v[0])
    vy.append(v[1])
plt.plot(vx,vy)

v=[-1,0]
vx=[v[0]]
vy=[v[1]]
for i in range(20*n):
    v=np.matmul(R,v)
    vx.append(v[0])
    vy.append(v[1])
plt.plot(vx,vy)

v=[-np.sqrt(1/2),-np.sqrt(1/2)]
vx=[v[0]]
vy=[v[1]]
for i in range(20*n):
    v=np.matmul(R,v)
    vx.append(v[0])
    vy.append(v[1])
plt.plot(vx,vy)

v=[0,-1]
vx=[v[0]]
vy=[v[1]]
for i in range(20*n):
    v=np.matmul(R,v)
    vx.append(v[0])
    vy.append(v[1])
plt.plot(vx,vy)


v=[np.sqrt(1/2),-np.sqrt(1/2)]
vx=[v[0]]
vy=[v[1]]
for i in range(20*n):
    v=np.matmul(R,v)
    vx.append(v[0])
    vy.append(v[1])
plt.plot(vx,vy)
d=plt.axis('off')
```
